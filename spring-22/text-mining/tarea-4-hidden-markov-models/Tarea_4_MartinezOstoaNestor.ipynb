#### Generación de Diccionarios

1. **Transiciones**

Calcula el número de veces que aparece un tag al lado de otro $$P(t_i|t_{i-1})$$

Las llaves son la tupla ```(prev_tag, tag)``` y los valores es la frecuencia en que esa tupla apareció 

2. **Emisiones**

Calcula la probabilidad de una palabra dado su tag $$P(w_i|t_i)$$

Las llaves son una tupla ```(tag, word)``` y los valores es la frecuencia de dicha tupla

3. **Tags**

Calcula las frecuencias de cada tag

````python
import numpy as np
import pandas as pd
from collections import defaultdict
from os import walk
import seaborn as sns
import matplotlib.pyplot as plt
import math

from brat_parser import get_entities_relations_attributes_groups
from sklearn.metrics import classification_report
````

````python
TRAINING_PATH = "training/"
TEST_PATH = "test/"
````

````python
def clean_word(word):
    return word.lower()

def build_corpus(root_path, clean=True):
    tags = []
    words = []
    for (dirpath, dirnames, filenames) in walk(root_path):
        for file in filenames:
            extension = file[-3:]
            if extension == "ann":
                path = root_path + "/" + file
                entities, relations, _, _ = get_entities_relations_attributes_groups(path)
                for k in entities.keys():
                    e = entities[k]
                    w = clean_word(e.type) if clean else e.type
                    t = clean_word(e.text) if clean else e.text
                    tags.append(w)
                    words.append(t)
    return pd.DataFrame({'word': words, 'tag': tags})
````

````python
training_df = build_corpus(TRAINING_PATH)
test_df = build_corpus(TEST_PATH)

print(f"Número de registros de entrenamiento: {training_df.shape[0]}")
print(f"Número de registros de prueba: {test_df.shape[0]}")

training_df.head()
````

````python
def build_vocab(training_corpus, test_corpus):
    training_words = training_corpus['word'].unique()
    test_words = test_corpus['word'].unique()
    vocab = dict()
    for i, word in enumerate(training_words):
        if word not in vocab:
            vocab[word] = i
    for i, word in enumerate(test_words):
        if word not in vocab:
            vocab[word] = i
    return vocab

def create_dictionaries(corpus, vocab):
    emission_counts = defaultdict(int)
    transition_counts = defaultdict(int)
    tag_counts = defaultdict(int)
    prev_tag = '--s--' 
    i = 0 
    for idx, row in corpus.iterrows():
        word = row['word']
        tag = row['tag']
        i += 1
        if word not in vocab:
            word = '--UNK--'
        transition_counts[(prev_tag, tag)] += 1
        emission_counts[(tag, word)] += 1
        tag_counts[tag] += 1
        prev_tag = tag
        
    return emission_counts, transition_counts, tag_counts
````

````python
vocab = build_vocab(training_df, test_df)
emission_counts, transition_counts, tag_counts = create_dictionaries(training_df, vocab)
states = sorted(tag_counts.keys())
````

#### Matriz de probabilidades de transición $A$

- En esta matriz, cada celda da la probabilidad de pasar de un ```tag``` a otro
- Las dimensiones de esta matriz son $(E,E)$ donde $E$ es el número de entidades (tags)

````python
def create_transition_matrix(alpha, tag_counts, transition_counts):
    all_tags = sorted(tag_counts.keys())
    num_tags = len(all_tags)
    A = np.zeros((num_tags,num_tags))
    trans_keys = set(transition_counts.keys())
    for i in range(num_tags):
        for j in range(num_tags):
            count = 0
            key = (all_tags[i], all_tags[j])
            if key in trans_keys: 
                count = transition_counts[key]
            count_prev_tag = tag_counts[all_tags[i]]
            A[i,j] = (count + alpha) / (count_prev_tag + alpha * num_tags)

    return A
````

````python
alpha = 0.001
A = create_transition_matrix(alpha, tag_counts, transition_counts)
pd.DataFrame(A[:,:], index=states, columns =states)
````

#### Matriz de probabilidades de emisión $B$

- Cada celda de la matriz $B$ indica la probabilidad de que cada palabra pertenezca a una de las entidades
- Dimensiones $(E, N)$ donde $E$ es el número de entidades y $N$ es el número de palabras únicas (longitud del vocabulario

````python
def create_emission_matrix(alpha, tag_counts, emission_counts, vocab):
    num_tags = len(tag_counts)
    all_tags = sorted(tag_counts.keys())
    num_words = len(vocab)
    B = np.zeros((num_tags, num_words))
    emis_keys = set(list(emission_counts.keys()))
    for i in range(num_tags): # complete this line
        for j in range(num_words): # complete this line
            count = 0
            key = (all_tags[i], vocab[j])
            if key in emis_keys: # complete this line
                count = emission_counts[key]
            count_tag = tag_counts[all_tags[i]]
            B[i,j] = (count + alpha) / (count_tag + alpha * num_words)
    return B
````

````python
B = create_emission_matrix(alpha, tag_counts, emission_counts, list(vocab))
cidx  = ['citracal','prograf','fentanyl', 'cream', '50 mg']
cols = [vocab[a] for a in cidx]
rows = [states.index(a) for a in states]
pd.DataFrame(B[np.ix_(rows,cols)], index=states, columns = cidx )
````

#### Algoritmo de Viterbi

````python
def initialize(states, tag_counts, A, B, corpus, vocab):
    num_tags = len(tag_counts)
    best_probs = np.zeros((num_tags, len(corpus)))
    best_paths = np.zeros((num_tags, len(corpus)), dtype=int)
    s_idx = 0
    for i in range(num_tags):
        if A[s_idx,i] == 0:
            best_probs[i,0] =  float('-inf')
        else:
            best_probs[i,0] = math.log(A[s_idx,i]) + math.log(B[i,vocab[corpus[0]]])
    return best_probs, best_paths
````

````python
prep = list(test_df['word'].values)
best_probs, best_paths = initialize(states, tag_counts, A, B, prep, vocab)
````

````python
def viterbi_forward(A, B, test_corpus, best_probs, best_paths, vocab):
    num_tags = best_probs.shape[0]
    for i in range(1, len(test_corpus)): 
        if i % 5000 == 0:
            print("Words processed: {:>8}".format(i))
        for j in range(num_tags):
            best_prob_i = float('-inf')
            best_path_i = None
            for k in range(num_tags): # complete this line
                prob = best_probs[k,i-1] + math.log(A[k,j]) + math.log(B[j,vocab[test_corpus[i]]])
                if prob > best_prob_i:
                    best_prob_i = prob
                    best_path_i = k
            best_probs[j,i] = best_prob_i
            best_paths[j,i] = best_path_i
    return best_probs, best_paths
````

````python
best_probs, best_paths = viterbi_forward(A, B, prep, best_probs, best_paths, vocab)
````

````python
def viterbi_backward(best_probs, best_paths, corpus, states):
    m = best_paths.shape[1] 
    z = [None] * m
    num_tags = best_probs.shape[0]
    best_prob_for_last_word = float('-inf')
    pred = [None] * m
    for k in range(num_tags): # complete this line
        if best_probs[k,-1] > best_prob_for_last_word: # complete this line
            best_prob_for_last_word = best_probs[k,-1]
            z[m - 1] = k
    pred[m - 1] = states[k]
    for i in range(len(corpus)-1,-1,-1): # complete this line
        pos_tag_for_word_i = best_paths[np.argmax(best_probs[:,i]),i]
        z[i - 1] = best_paths[pos_tag_for_word_i,i]
        pred[i - 1] = states[pos_tag_for_word_i]
    return pred
````

````python
pred = viterbi_backward(best_probs, best_paths, prep, states)
````

### Predicciones

````python
y_true = test_df['tag'].values
y_pred = np.array(pred)

print(classification_report(y_true, y_pred, digits=4))
````

````python
import pandas as pd
import numpy as np

import sklearn_crfsuite
from sklearn_crfsuite import scorers
from sklearn_crfsuite import metrics

import scipy.stats
from sklearn.metrics import make_scorer
from sklearn.model_selection import RandomizedSearchCV

import warnings
warnings.filterwarnings("ignore")

import random
````

````python
def build_corpus(root_path, clean=True):
    sents = []
    for (dirpath, dirnames, filenames) in walk(root_path):
        for file in filenames:
            extension = file[-3:]
            if extension == "ann":
                path = root_path + "/" + file
                entities, relations, _, _ = get_entities_relations_attributes_groups(path)
                tags = []
                words = []
                for k in entities.keys():
                    e = entities[k]
                    w = clean_word(e.type) if clean else e.type
                    t = clean_word(e.text) if clean else e.text
                    tags.append(w)
                    words.append(t)
                sents.append(list(zip(words, tags)))
    return sents
````

````python
training_sents = build_corpus(TRAINING_PATH, clean=False)
test_sents = build_corpus(TEST_PATH, clean=False)
````

````python
print(len(training_sents), len(test_sents))
````

````python
def word2features(sent, i):
    word = sent[i][0]
    postag = sent[i][1]
    
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(),
        'word[-3:]': word[-3:],
        'word[-2:]': word[-2:],
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        'postag': postag,
        'postag[:2]': postag[:2],
    }
    if i > 0:
        word1 = sent[i-1][0]
        postag1 = sent[i-1][1]
        features.update({
            '-1:word.lower()': word1.lower(),
            '-1:word.istitle()': word1.istitle(),
            '-1:word.isupper()': word1.isupper(),
            '-1:postag': postag1,
            '-1:postag[:2]': postag1[:2],
        })
    else:
        features['BOS'] = True

    if i < len(sent)-1:
        word1 = sent[i+1][0]
        postag1 = sent[i+1][1]
        features.update({
            '+1:word.lower()': word1.lower(),
            '+1:word.istitle()': word1.istitle(),
            '+1:word.isupper()': word1.isupper(),
            '+1:postag': postag1,
            '+1:postag[:2]': postag1[:2],
        })
    else:
        features['EOS'] = True
        
    return features


def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]

def sent2labels(sent):
    return [label for token, label in sent]
````

````python
X_train = [sent2features(s) for s in training_sents]
y_train = [sent2labels(s) for s in training_sents]

X_test = [sent2features(s) for s in test_sents]
y_test = [sent2labels(s) for s in test_sents]
````

````python
crf = sklearn_crfsuite.CRF(
    algorithm='l2sgd',
    c2=0.1,
    all_possible_transitions=True
)
crf.fit(X_train, y_train)
````

````python
labels = crf.classes_
y_pred = crf.predict(X_test)

y_true_flat = [e for l in y_test for e in l]
y_pred_flat = [e for l in y_pred for e in l]

print(classification_report(y_true_flat, y_pred_flat, digits=4))
````
