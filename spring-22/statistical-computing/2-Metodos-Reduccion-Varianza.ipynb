# Métodos de reducción de varianza

```python
from random import seed
import numpy as np
from numpy.random import random
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
seed(34)
```

```python
def funcion(x,a):
    return np.exp(x**a)
```

```python
valor_real = quad(funcion, 0, 1, 2)
```

```python
valor_real[0]
```

```python
def integracion_MonteCarlo(n, funcion, *args):
    muestra = 0
    for i in range(2*N):
        u = random()
        muestra += funcion(u, *args)
    return muestra/(2*N)
```

```python
N = 1000
MonteCarlo = np.zeros(N)
```

```python
for i in range(N):
    MonteCarlo[i] = integracion_MonteCarlo(i, funcion, 2)
```

```python
plt.plot(MonteCarlo, color="b", label="Monte Carlo crudo")
plt.axhline(y=valor_real[0], color='r', linestyle='--')
plt.legend()
plt.show()
```
