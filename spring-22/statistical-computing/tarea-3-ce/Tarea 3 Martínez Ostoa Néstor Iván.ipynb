## Ejercicio 1 - Algoritmo EM para una mixtura de Poisson

### Función de probabilidad para Poisson:

$$PMF = \frac{\lambda^k e^{-\lambda}}{k!}$$

```python
def pmf(lambda_, k):
    return (lambda_**k) * np.exp(-lambda_) / np.math.factorial(k)
```

### Probabilidad posterior

$$A_{i,k} = P(Z_i = k| X_i, \theta^0) $$ 

---

$$A_{i,k} = \frac{P(X_i | Z_i = k, \theta^0)\pi_k}{\sum_{k'=1}^K P(X_i | Z_i = k', \theta^0)\pi_{k'}} $$

```python
def posterior(X, lambdas, probs):
    posterior_prob = []
    for xi in X:        
        cluster = []
        for i in range(len(lambdas)):
            cluster.append(pmf(lambdas[i], xi) * probs[i])
        posterior = [c/sum(cluster) for c in cluster]
        posterior_prob.append(posterior)
    return np.array(posterior_prob).T.tolist()
```

### Maximización - encontrando $\hat{\pi}_1$ y $\hat{\pi}_2$

$$\hat{\pi}_1 = \frac{\sum_{i=1}^NA_{i,1}}{\sum_{i=1}^NA_{i,1} + A_{i,2}}$$

---

$$\hat{\pi}_2 = 1 - \hat{\pi}_1$$

```python
def find_pi(posterior):
    a1 = posterior[0]
    a2 = posterior[1]
    a1a2 = [a + b for a, b in zip(a1, a2)]
    pi1 = sum(a1) / sum(a1a2)
    return pi1, 1-pi1
```

### Maximización - encontrando $\lambda$

$$\hat{\lambda}_k = \frac{\sum_{i=1}^N A_{i,k}X_i}{\sum_{i=1}^N A_{i,k}} \quad \text{para cada cluster } k$$

```python
def find_lambdas(X, posterior):
    lambdas = []
    for p in posterior:
        lambda_x_post = [a*b for a, b in zip(X, p)]
        lambdas.append(sum(lambda_x_post) / sum(p))
    return lambdas
```

### Algoritmo EM

1. Inicialización de parámetros aleatorios
2. Obtener la esperanza de la log verosimilitud completa $Q(\theta, \theta^0)$
3. Calcular probabilidades posteriores 
4. Encontrar parámetros óptimos diferenciando $Q(\theta, \theta^0)$ con respecto a cada parámetro e igualando a cero
5. Actualizar parámetros hasta criterio de aceptación

```python
def EM(X, lambdas, probs):
    incomplete_log_likelihoods = []
    delta = 1
    eps = 1e-5
    
    while delta > eps:
        posteriors = posterior(X, lambdas, probs)
        new_probs = find_pi(posteriors)
        new_lambdas = find_lambdas(X, posteriors)
        
        ill = incomplete_log_likelihood(X, new_lambdas, new_probs)
        incomplete_log_likelihoods.append(ill)
        
        if len(incomplete_log_likelihoods) > 1:
            delta = incomplete_log_likelihoods[-1] - incomplete_log_likelihoods[-2]
    
    return new_lambdas, new_probs
```

## Ejercicio 2 - Robins-Monro

Implemente el esquema de Robbins-Monro para elegir el hiper parámetro de desviación estándar en la propuesta de caminata aleatoria del algoritmo de Metrópolis-Hastings con una probabilidad de aceptación de objetivo $\alpha = 0.5$. Ilustre que el algoritmo está bien implementado

```python
def P(x):
    return beta.pdf(x, 0.35, 0.65)
```

```python
samples = []
x = 0.2
n = 100000
probs = []
sigma = 1
alpha = 0.5
success = 0

for i in range(1, n+1):
    y = x + np.random.normal(0, sigma**2)
    prob_acceptance = min(1, P(y) / P(x))
    if (np.random.uniform(0, 1) < prob_acceptance):
        x = y
        sigma += (1-alpha)/i
        success += 1
    else:
        sigma -= alpha/i
    samples.append(x)
    probs.append(prob_acceptance)
    
samples = np.array(samples)
```

```python
print(f"n = {n}\nexitos = {success}\nProporción de éxitos = {success/n}")
```

```python
print(f"sigma = {sigma}")
```

Dado que queremos que nuestra probabilidad de aceptación sea del $0.5$, podemos observar que nuestra proporción de éxitos es muy cercana a $0.5$ por lo que el algoritmo queda justificado

```python
fig, ax = plt.subplots(1, 1)

sns.histplot(
    samples+7, stat='density', bins=20, palette="light:m_r",
    linewidth=1, legend=False
)

X = np.linspace(0.01, 0.99, n)
Y = [P(i) for i in X]
sns.lineplot(
    x=X+7, y=Y, palette="light:m_r", legend=False
)

ax.set_xlabel('x')
ax.set_ylabel('PDF')
ax.set_title(f"Metropolis-Hastings para X = beta(0.35, 0.65) + 7\nProbabilidad de aceptación: {np.round(np.mean(probs), 4)}")
plt.legend(labels=['X - Teórico','Muestras de X - MH'])
plt.show()
```
