## Ejercicio 1

Programe el algoritmo de Metrópolis-Hastings (MH) basado en una caminata aleatoria
Gaussiana para la distribución de propuesta en el lenguaje de programación de su
preferencia para simular $X = Z + 7$ tal que $Z$ tiene una distribución Beta$(0.35, 0.65)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from scipy.special import gamma, factorial
import seaborn as sns
import pandas as pd
sns.set_theme(style="darkgrid")
```

```python
n = 10000
```

```python
def P(x):
    return beta.pdf(x, 0.35, 0.65)
```

```python
X = np.linspace(0.01, 0.99, n)
Y = [P(i) for i in X]
fig, ax = plt.subplots(1,1)
ax.plot(X+7, Y, '-r')
ax.set_ylim([0,2])
plt.show()
```

```python
samples = []
x = 0.2
probs = []
for i in range(n):
    y = x + np.random.normal(size=1)[0]
    prob_acceptance = min(1, P(y) / P(x))
    if (np.random.uniform(0, 1) < prob_acceptance):
        x = y
    samples.append(x)
    probs.append(prob_acceptance)
    
samples = np.array(samples)
```

```python
fig, ax = plt.subplots(1, 1)

sns.histplot(
    samples+7, stat='density', bins=20, palette="light:m_r",
    linewidth=1, legend=False
)

sns.lineplot(
    x=X+7, y=Y, palette="light:m_r", legend=False
)

ax.set_xlabel('x')
ax.set_ylabel('PDF')
ax.set_title(f"Metropolis-Hastings para X = beta(0.35, 0.65) + 7\nProbabilidad de aceptación: {np.round(np.mean(probs), 4)}")
plt.legend(labels=['X - Teórico','Muestras de X - MH'])
plt.savefig('Ejercicio1.png', format='png')
plt.show()
```
