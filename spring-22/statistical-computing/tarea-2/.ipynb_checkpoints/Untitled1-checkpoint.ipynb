```r
gamm <- function(n, a, b) {
    mu <- a/b
    sig <- sqrt(a/(b*b))
    vec <- vector("numeric", n)
    x <- a/b
    vec[1] <- x
    for (i in 2:n) {
        candidate <- rnorm(1, mu, sig)
        accept <- min(
            1,
            (dgamma(candidate, a, b) / dgamma(x, a, b)) / (dnorm(candidate, mu, sig))
        )
        u <- runif(1)
        if (u < accept)
            x <- candidate
        vec[i] <- candidate
    }
    vec
}
```

```r
vec <- gamm(100000, 2.3, 2.7)
```

```r
hist(vec, 30)
```

```r
c(mean(vec), var(vec))
```
