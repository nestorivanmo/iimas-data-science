# Simulación Montecarlo

- Estimar áreas a través de distribuciones
- Estimación de $\pi$

```python
from random import seed
from numpy.random import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
seed(92)
```

```python
def estimar_pi(N):
    ensayos = N
    exito = 0
    for i in range(ensayos):
        horizontal = (random() * 2) - 1
        vertical = (random() * 2) - 1
        if (horizontal ** 2 + vertical ** 2) ** 0.5 < 1:
            exito += 1
    pi_estimado = 4*(exito / ensayos)
    return pi_estimado
```

```python
for i in np.linspace(1,1e5,5):
    print(f"i={i}\tpi={estimar_pi(int(i))}")
```

```python
n = 1000
vec = np.zeros(n)
```

```python
for i in range(n):
    vec[i] = estimar_pi(i+1)
```

```python
plt.plot(vec, color="b")
plt.axhline(y = np.pi, color='r', linestyle = '--')
plt.show()
```

```python
def lanzamientos(N):
    ensayos = N
    exitos = 0
    h, v = np.zeros(N), np.zeros(N)
    col = []
    for i in range(ensayos):
        h[i] = (random() * 2) - 1
        v[i] = (random() * 2) - 1
        if (h[i]**2 + v[i]**2)**0.5 < 1:
            col.append("blue")
            exitos += 1
        else:
            col.append("magenta")
    pi_estimado = 4*(exitos / ensayos)
    return pi_estimado, [h, v], col
            
```

```python
n = 1000
pi, lanz, col = lanzamientos(n)
```

```python
plt.axes()
plt.scatter(lanz[0], lanz[1], c=col)
plt.axis('scaled')
plt.show()
```

## Aguja de Buffon

$$\pi \approx \frac{2lN}{Kd}$$ donde:
- $l$ es la longitud de la aguja
- $N$ es el número de ensayos
- $K$ es el número de éxitos
- $d$ es la distancia de la aguja a la línea paralela más cercana

```python
def aguja_buffon(N, lineas, l, d):
    if d < l:
        raise Exception('l debe er menor o igual a d')
    lin_lim = lineas*d - 1
    cuadricula = np.linspace(0, lin_lim, lineas)
    exito = 0
    for i in range(N):
        x =random()*lin_lim
        x_dist = min(abs(x-cuadricula))
        theta = random()*np.pi*0.5
        if (x_dist < (l/2)*np.sin(theta)):
            exito += 1
    if (exito == 0):
        return 0
    return (2*l*N)/(exito*d)
```

```python
pi_aguja = aguja_buffon(10000, 7, 1, 1)
pi_aguja
```

```python
n = 5000
vec2 = np.zeros(n)
```

```python
for i in range(1, n):
    vec2[i] = aguja_buffon(i, 5, 1, 1)
```

```python
plt.plot(vec2, color='g')
plt.axhline(y=np.pi, color='r', linestyle='--')
plt.show()
```
