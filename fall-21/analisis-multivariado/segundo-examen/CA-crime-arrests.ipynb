## Bibliotecas requeridas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

## Datos

```python
crime_data = pd.read_csv('crime.dat', sep='\s+')
crime_data = crime_data.iloc[:, :-2]
print(crime_data.shape)
crime_data.head()
```

```python
def build_data():
    crime_data = pd.read_csv('crime.dat', sep='\s+')
    crime_data = crime_data.iloc[:, :-2]
    crime_data = crime_data.iloc[:, 1:].values
    scaled_data = scaler.fit_transform(crime_data)
    return scaled_data

def plot_data(method, name):
    scaled_data = build_data()
    plt.figure(figsize=(20,5))
    plt.title(f"Crímenes en E.U.A. - {name}")
    dend = sch.dendrogram(sch.linkage(scaled_data, method=method))
    plt.xlabel('Tazas de crimen por estado')
    plt.ylabel('Distancias')
    plt.savefig(f'imgs/{method}_crime_data.png', dpi=150)
```

## Single Linkage

"Tends to produce unbalanced and straggly clusters (‘chaining’), especially in large data sets. Does not take account of cluster structure."

```python
plot_data(method='single', name='Single Linkage')
```

## Complete Linkage

```python
plot_data(method='complete', name='Complete Linkage')
```

## Group Average Linkage - UPGMA

```python
plot_data(method='average', name='Average Linkage')
```

## Centroid Linkage - UPGMC

```python
crime_data = pd.read_csv('crime.dat', sep='\s+')
crime_data = crime_data.iloc[:, :-2]
crime_data = crime_data.iloc[:, 1:].values

plt.figure(figsize=(20,5))
plt.title(f"Crímenes en E.U.A. - Centroid Linkage")
method = 'centroid'
dend = sch.dendrogram(sch.linkage(crime_data, method=method))
plt.xlabel('Tazas de crimen por estado')
plt.ylabel('Distancias')
plt.savefig(f'imgs/{method}_crime_data.png', dpi=150)
```

## Weighted Average Linkage  - WPGMA

```python
plot_data(method='weighted', name='Weighted Linkage')
```
