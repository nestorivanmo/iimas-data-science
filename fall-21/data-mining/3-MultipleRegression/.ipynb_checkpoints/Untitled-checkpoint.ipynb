```r
library("rpart")
library("rpart.plot")
library("caret")
library("caTools")
```

```r
setwd(getwd())
df <- read.csv("data.csv")
View(df)
```

```r
# selected_data <- data[c('carlength','stroke','enginesize','boreratio','cylindernumber','compressionratio','horsepower','aspiration','peakrpm','carbody','drivewheel','highwaympg','citympg','curbweight')]
selected_data <- df[c('enginesize', 'horsepower', 'carwidth', 'curbweight', 'price')]
```

```r
tree1 = rpart(
    selected_data$price ~ .,
    data=selected_data
)
tree1
```

```r
rpart.plot(tree1)
```
